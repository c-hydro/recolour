### **Configure the processed datasets**

#### Set the time-series processed datasets

```python
# Define the folder name and the file name
file_name_ts_processed_def = file_name_ts_processed_raw.format(point_name = info_point_value)

path_name_processed_ts = os.path.join(folder_name_ts_processed, file_name_ts_processed_def)

label_reference = 'ts_reference'
label_other = 'ts_other'

# Info
print(' ==> Define the datasets file name: "' + path_name_processed_ts + '" ... DONE')
```

#### Load the datasets

```python
# Load the datasets in pickle format
dframe_ts_processed = read_obj(path_name_processed_ts)

# Info
print(dframe_ts_processed.head(5))
print(' ==> Load the time-series file name ... DONE')
```

### **Analyze the processed datasets**

#### Seasonal datasets

```python
# Load the seasons
info_season_list = sorted(list(set(list(dframe_ts_processed['season'].values) + ['ALL'])))
```

```python
# Display available outlet sections
info_season_select = widgets.Dropdown(options=info_season_list, description='Season', disabled=False, rows=5)
display(info_season_select)  
```

```python
# Parser the selection from the scroll menu
info_season_value = str(info_season_select.value)

# Info
print(' ==> Select the season name "' + info_season_value + '" ... DONE')
```

```python
# Get the seasonal dframe
if info_season_value == 'ALL':
    dframe_ts_filtered = deepcopy(dframe_ts_processed)
else:
    dframe_ts_filtered = dframe_ts_processed.loc[dframe_ts_processed['season'] == info_season_value]
    
# Info
print(dframe_ts_filtered.head(5))
print(' ==> Select the season datasets ... DONE')
```

```python
# Drop the seasonal column
dframe_ts_filtered = dframe_ts_filtered.drop(columns=['season'])

# Info
print(dframe_ts_filtered.head(5))
print(' ==> Select the season datasets ... DONE')
```

#### Scale datasets

```python
# Scale datasets using a method defined in pytestmo library
dframe_ts_scaled = scaling.scale(dframe_ts_filtered, method='cdf_beta_match', reference_index=1)

# Info
print(' ==> Scale the datasets ... DONE')
```

#### Compute metrics

```python
# Set the name of the plot
info_graph_value = 'metrics_scaled'
```

```python
# Calculate correlation coefficients, RMSD, bias, Nash Sutcliffe
x, y = dframe_ts_scaled[label_reference].values, dframe_ts_scaled[label_other].values

metrics_list = []
metrics_list.append("Pearson's R    = {:.2f}, p = {:.2e}".format(*stats.pearsonr(x, y)))
metrics_list.append("Spearman's rho = {:.2f}, p = {:.2e}".format(*stats.spearmanr(x, y)))
metrics_list.append("Kendall's tau  = {:.2f}, p = {:.2e}".format(*stats.kendalltau(x, y)))
metrics_list.append('')
metrics_list.append("RMSD = {:.2f}".format(metrics.rmsd(x, y)))
metrics_list.append("Bias = {:.2f}".format(metrics.bias(x, y)))
metrics_list.append("Nash Sutcliffe = {:.2f}".format(metrics.nash_sutcliffe(x, y)))

# Info
print(' ==> Compute the metrics ... DONE')
```

```python
# View the computed metrics
for metrics_step in metrics_list:
    print(' ===> ' + metrics_step)

# Info
print(' ==> View the metrics ... DONE')
```

```python
# Create file name
info_metrics_dict = {'graph_name': info_graph_value, 'point_name': info_point_value, 'season_name': info_season_value}
file_name_metrics_def = file_name_metrics_raw.format(**info_metrics_dict)

# Join the folder and the file name(s)
file_path_metrics = os.path.join(folder_name_metrics, file_name_metrics_def)

# Make the image folder
if not os.path.exists(folder_name_metrics):
    os.makedirs(folder_name_metrics)

# Info
print(' ==> Define the metrics file name "' + file_path_metrics + '" ... DONE')
```

```python
# Dump the metrics to the file
with open(file_path_metrics, 'a') as file_handle:
    for metrics_step in metrics_list: 
        file_handle.write(metrics_step + '\n')

# Info
print(' ==> Dump the metrics data ... DONE')
```

#### Plot the time-series

```python
# Set the name of the plot
info_graph_value = 'ts_scaled'
file_dpi_image = 150
```

```python
# Plot the datasets
fig, ax = plt.subplots()

#dframe_ts_scaled = dframe_ts_scaled.fillna(value=np.nan)

dframe_ts_scaled.plot(figsize=(10, 5), ax=ax)
ax.set_xlabel('time')
ax.set_ylabel('sm [-]')
ax.legend(["sm modelled", "sm observed"])
ax.grid(True)

plt.ylim(0, 60)
plt.title('sm - modelled vs observed - season ' + info_season_value)
plt.show()

# Info
print(' ==> Plot the time-series figure ... DONE')
```

```python
# Create plot name
info_image_dict = {'graph_name': info_graph_value, 'point_name': info_point_value, 'season_name': info_season_value}
file_name_image_def = file_name_image_raw.format(**info_image_dict)

# Join the folder and the file name(s)
file_path_image = os.path.join(folder_name_image, file_name_image_def)

# Make the image folder
if not os.path.exists(folder_name_image):
    os.makedirs(folder_name_image)

# Info
print(' ==> Define the time-series file name "' + file_path_image + '" ... DONE')
```

```python
# Dump the plot
fig.savefig(file_path_image, dpi=file_dpi_image)

# Info
print(' ==> Dump the time-series figure ... DONE')
```

#### Plot the scatter

```python
# Set the name of the plot
info_graph_value = 'scatter_scaled'
file_dpi_image = 150
```

```python
# Plot the datasets
fig, ax = plt.subplots()

ax.scatter(dframe_ts_scaled[label_reference].values, dframe_ts_scaled[label_other].values, color='purple')
ax.set_xlabel('sm modelled')
ax.set_ylabel('sm observed')
ax.grid(True)

# Fit linear regression via least squares with numpy.polyfit
# It returns an slope (b) and intercept (a)
# deg=1 means linear fit (i.e. polynomial of degree 1)
b, a = np.polyfit(dframe_ts_scaled[label_reference].values, dframe_ts_scaled[label_other].values, deg=1)

# Create sequence of 100 numbers from 0 to 100 
xseq = np.linspace(0, 100, num=100)

# Plot regression line
ax.plot(xseq, a + b * xseq, lw=2.5, color='steelblue', linestyle='--', linewidth=2);

sns.regplot(x=dframe_ts_scaled[label_reference].values, y=dframe_ts_scaled[label_other].values)

ax.plot([0, 25, 50, 75, 100], [0, 25, 50, 75, 100], color='k', linestyle='-', linewidth=0.2)

plt.ylim(0, 60);
plt.xlim(0, 60);
plt.title('sm - modelled vs observed - season ' + info_season_value)

# Info
print(' ==> Plot the scatter figure ... DONE')
```

```python
# Create plot name
info_image_dict = {'graph_name': info_graph_value, 'point_name': info_point_value, 'season_name': info_season_value}
file_name_image_def = file_name_image_raw.format(**info_image_dict)

# Join the folder and the file name(s)
file_path_image = os.path.join(folder_name_image, file_name_image_def)

# Make the image folder
if not os.path.exists(folder_name_image):
    os.makedirs(folder_name_image)

# Info
print(' ==> Define the scatter file name "' + file_path_image + '" ... DONE')
```

```python
# Dump the plot
fig.savefig(file_path_image, dpi=file_dpi_image)

# Info
print(' ==> Dump the scatter figure ... DONE')
```
